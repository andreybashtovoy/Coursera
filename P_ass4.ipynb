```python
##загружаем библиотеки
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
import random
```

Используем распределение Парето https://ru.wikipedia.org/wiki/%D0%A0%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5_%D0%9F%D0%B0%D1%80%D0%B5%D1%82%D0%BE

```python
a, m = 3., 2.  # shape and mode
s = np.random.pareto(a, 1000)
```

```python
## Теоретические значения мат.ожидания и дисперсии
mu = a*m**a/m**(a+1) #a*m/(a-1) #np.mean(s) #
sig2 = (m/(a-1))**2*(a/(a-2))
print mu, sig2
```

```python
s1 = (np.random.pareto(a, 1000) + 1) * m
count, bins, _ = plt.hist(s1, 100, normed=True)
fit = a*m**a / bins**(a+1)
plt.plot(bins, max(count)*fit/max(fit), label='pareto', linewidth=2, color='r')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
## 1000 выборок длиной 5
n = 5
m_rnd = [None] * 1000
for k in range(0,1000,1):
    rand_smpl = np.random.pareto(a, n)  #[ s[i] for i in sorted(random.sample(xrange(len(s)), n)) ]
    m_rnd[k]=np.mean(rand_smpl)
# m_rnd содержит среднее значение выборки
count, bins, _ = plt.hist(m_rnd, 100, normed=True, label='our pdf')
#параметры для нормального распределения
#mu5= np.mean(m_rnd)
#sig5=np.std(m_rnd)
norm_rv = sts.norm(mu, (sig2/n)**0.5 ) #sts.norm(mu5, sig5)
sample = norm_rv.rvs(100)

# на том же графике построим теоретическую плотность распределения:
x = np.linspace(0,5,100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5,  color='r')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```
