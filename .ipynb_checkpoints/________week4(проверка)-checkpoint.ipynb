# Равномерное распределение
## Подключение библиотек

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts

%matplotlib inline
```

## Генерация выборки

```python
unif_rv = sts.uniform
my_arr = unif_rv.rvs(size=1000)
```

### Строим распределение

```python
x = np.linspace(-1,2,1000)
cdf = unif_rv.cdf(x)
plt.plot(x, cdf)
```

### Строим гистограмму и плотность

```python
#Плотность
df = pd.DataFrame(my_arr, columns=['KDE'])
ax = df.plot(kind='density')
#Гистограмма
plt.hist(my_arr, bins=40, normed = True)
plt.ylabel('number of samples')
plt.xlabel('$x$')

```

## Генерация множества выборок

```python
unif_rv = sts.uniform
x = np.linspace(-1,2,1000)
```

# Выборки размера 5

```python
n=5
av = []
c=1
while c <=1000:
    my_arr = unif_rv.rvs(size=n)
    av.append(np.average(my_arr))
    c+=1
#Нормальное распределение
norm_rv = sts.norm(0.5, 0.129099445)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
#гистограмма
plt.hist(av, normed=True)
plt.ylabel('number of samples')
plt.xlabel('$x$')
```

# Выборки размера 10

```python
n=10
av = []
c=1
while c <=1000:
    my_arr = unif_rv.rvs(size=n)
    av.append(np.average(my_arr))
    c+=1
#Нормальное распределение
norm_rv = sts.norm(0.5, 0.091287093)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
#гистограмма
plt.hist(av, normed=True)
plt.ylabel('number of samples')
plt.xlabel('$x$')
```

# Выборки размера 50

```python
n=50
av = []
c=1
while c <=1000:
    my_arr = unif_rv.rvs(size=n)
    av.append(np.average(my_arr))
    c+=1
#Нормальное распределение
norm_rv = sts.norm(0.5, 0.040824829)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
#гистограмма
plt.hist(av, normed=True)
plt.ylabel('number of samples')
plt.xlabel('$x$')
```

# Выводы
Точность апроксимации увеличивается с размером выборки, хотя для небольших размеров(5 и 10) разница в качестве апроксимации не особа различима.
